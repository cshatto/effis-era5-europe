```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import os

# Ensure figures directory exists
os.makedirs('figures', exist_ok=True)

# Use filtered_df from previous script (AREA_HA <= 2500)
# Assuming it has 'FWI', 'vpd', 'AREA_HA'

# Function to remove outliers using IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers from AREA_HA, FWI, and vpd
df_clean = filtered_df.copy()

# df_clean = remove_outliers(df_clean, 'AREA_HA')
# df_clean = remove_outliers(df_clean, 'FWI')
# df_clean = remove_outliers(df_clean, 'vpd')
print(f"Original data points: {len(filtered_df)}, After outlier removal: {len(df_clean)}")

# Remove NaN values
df_clean = df_clean.dropna(subset=['AREA_HA', 'FWI', 'vpd'])

# --- 1. Correlation Analysis ---
correlation_fwi = df_clean['FWI'].corr(df_clean['AREA_HA'])
correlation_vpd = df_clean['vpd'].corr(df_clean['AREA_HA'])
print(f"Correlation between FWI and AREA_HA: {correlation_fwi:.3f}")
print(f"Correlation between vpd and AREA_HA: {correlation_vpd:.3f}")

# --- 2. Simple Linear Regression ---
# Prepare data
X_fwi = df_clean[['FWI']].values
X_vpd = df_clean[['vpd']].values
y = df_clean['AREA_HA'].values

# Fit models
model_fwi = LinearRegression()
model_fwi.fit(X_fwi, y)
y_pred_fwi = model_fwi.predict(X_fwi)

model_vpd = LinearRegression()
model_vpd.fit(X_vpd, y)
y_pred_vpd = model_vpd.predict(X_vpd)

# Evaluate models
r2_fwi = r2_score(y, y_pred_fwi)
rmse_fwi = np.sqrt(mean_squared_error(y, y_pred_fwi))
r2_vpd = r2_score(y, y_pred_vpd)
rmse_vpd = np.sqrt(mean_squared_error(y, y_pred_vpd))

print("\nSimple Linear Regression Results (outliers removed):")
print(f"FWI: R^2 = {r2_fwi:.3f}, RMSE = {rmse_fwi:.3f}")
print(f"vpd: R^2 = {r2_vpd:.3f}, RMSE = {rmse_vpd:.3f}")

# --- 3. Multiple Linear Regression ---
X_combined = df_clean[['FWI', 'vpd']].values
model_combined = LinearRegression()
model_combined.fit(X_combined, y)
y_pred_combined = model_combined.predict(X_combined)

r2_combined = r2_score(y, y_pred_combined)
rmse_combined = np.sqrt(mean_squared_error(y, y_pred_combined))
print("\nMultiple Linear Regression (FWI + vpd, outliers removed):")
print(f"R^2 = {r2_combined:.3f}, RMSE = {rmse_combined:.3f}")
print(f"Coefficients: FWI = {model_combined.coef_[0]:.3f}, vpd = {model_combined.coef_[1]:.3f}")
print(f"Intercept: {model_combined.intercept_:.3f}")

# --- 4. Visualization ---
# Set Seaborn style
sns.set(style="whitegrid")

# Scatter plot: FWI vs AREA_HA
plt.figure(figsize=(10, 6))
sns.scatterplot(x='FWI', y='AREA_HA', data=df_clean, alpha=0.5)
plt.plot(df_clean['FWI'], y_pred_fwi, color='red', label=f'R^2 = {r2_fwi:.3f}')
plt.title('AREA_HA vs FWI with Regression Line (Outliers Removed)')
plt.xlabel('Fire Weather Index (FWI)')
plt.ylabel('Area (ha)')
plt.legend()
plt.tight_layout()
plt.savefig('figures/area_ha_vs_fwi_regression_no_outliers.png')
plt.close()

# Scatter plot: vpd vs AREA_HA
plt.figure(figsize=(10, 6))
sns.scatterplot(x='vpd', y='AREA_HA', data=df_clean, alpha=0.5)
plt.plot(df_clean['vpd'], y_pred_vpd, color='red', label=f'R^2 = {r2_vpd:.3f}')
plt.title('AREA_HA vs vpd with Regression Line (Outliers Removed)')
plt.xlabel('Vapor Pressure Deficit (vpd)')
plt.ylabel('Area (ha)')
plt.legend()
plt.tight_layout()
plt.savefig('figures/area_ha_vs_vpd_regression_no_outliers.png')
plt.close()

print("Regression plots (without outliers) saved in 'figures/' folder.")
```
